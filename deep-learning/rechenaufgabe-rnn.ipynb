- Input $x$
- Hidden State $h$
- Output $o$
- Initialem Hidden State $h_0 = (0,1,1)^T$

Als Aktivierungsfunktion werde $h^\text{relu}$ genutzt.

Die Gewichtsmatrizen seien gegeben durch
